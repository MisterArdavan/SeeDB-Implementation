```python
import pandas as pd
```

```python
df = pd.DataFrame(pd.read_csv('adultdata_dirty.csv', na_values=" ?"))
string_cols = ["WORK_CLASS", "EDUCATION", "MARITAL_STATUS", "OCCUPATION", "RELATIONSHIP", "RACE", "SEX", "NATIVE_COUNTRY", "CLASS"]

#remove row if it contains ?
df = df.dropna(axis=0)

for col in string_cols:
  #remove white spaces
  df[col] = df[col].str.strip()

df.to_csv('adultdata.csv')  
```
